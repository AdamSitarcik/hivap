```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
proj = '88Sr'
trg = '92Mo'
cn = '180Hg'
bf = 0.85
ch = 'pxn'

file_name10 = proj + '_' + trg + '_' + cn + '_' + ch + '_' + 'IFUS10' + '_' + 'barfac' + str(bf) + '.dat'
file_name0 = proj + '_' + trg + '_' + cn + '_' + ch + '_' + 'IFUS0' + '_' + 'barfac' + str(bf) + '.dat'
print(file_name10, file_name0)
```

```python
pxn_10 = open('data/' + file_name10, 'r')
pxn_0 = open('data/' + file_name0, 'r')
```

```python
df_pxn10 = pd.read_csv('data/' + file_name10, sep='\t')
df_pxn0 = pd.read_csv('data/' + file_name0, sep='\t')
```

```python
fig, ax = plt.subplots(figsize=(15,10))
type(ax)
ax.plot(df_pxn10.iloc[:, 0], df_pxn10.iloc[:, 5])
ax.semilogy()
ax.set_xlabel('$E_{proj}$ [MeV]')
ax.set_ylabel('$\sigma$ [mb]')
ax.legend(df_pxn10.columns[2:], loc='upper right')
ax.set_ylim([1e-3,1e1])


ax.plot(df_pxn0.iloc[:, 0], df_pxn0.iloc[:, 5], '--')
# ax.legend(df_pxn0.columns[2:], loc='lower right')
```
