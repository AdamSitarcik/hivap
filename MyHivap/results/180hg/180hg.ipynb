```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pxn_10 = open('data/88Sr_92Mo_180Hg_pxn_IFUS10_barfac0.85.dat', 'r')
pxn_0 = open('data/88Sr_92Mo_180Hg_pxn_IFUS0_barfac0.85.dat', 'r')
```

```python
df_pxn10 = pd.read_csv('data/88Sr_92Mo_180Hg_pxn_IFUS10_barfac0.85.dat', sep='\t')
df_pxn0 = pd.read_csv('data/88Sr_92Mo_180Hg_pxn_IFUS0_barfac0.85.dat', sep='\t')
```

```python
fig, ax = plt.subplots(figsize=(15,10))
type(ax)
ax.plot(df_pxn10.iloc[:, 0], df_pxn10.iloc[:, 5])
ax.semilogy()
ax.set_xlabel('$E_{proj}$ [MeV]')
ax.set_ylabel('$\sigma$ [mb]')
ax.legend(df_pxn10.columns[2:], loc='upper right')
ax.set_ylim([1e-3,1e1])


ax.plot(df_pxn0.iloc[:, 0], df_pxn0.iloc[:, 5], '--')
# ax.legend(df_pxn0.columns[2:], loc='lower right')
```
