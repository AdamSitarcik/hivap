```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.lines import Line2D
```

```python
proj = '40Ar'
trg = '165Ho'
cn = '205At'
bf = 0.81
ch = 'xn'

sigr_dict = {}
# sigr_array = np.arange(2.5, 3.5, 0.4)
sigr_array = np.array([1.5, 2.5, 3.3, 5.0])
# sigr_array = np.array([2.5, 3.3])
for sigr in sigr_array:
    sigr_value = str(f'{sigr:.1f}')
    sigr_string = 'sigr' + sigr_value
    
    df10_mb = pd.read_csv('data/' + proj + '_' + trg + '_' + cn + '_' + ch + '_' + 'IFUS10' + '_barfac' + f'{bf:.2f}' + '_sigr' + sigr_value + '.dat', sep='\t')
    
    dfm10_mb = df10_mb.melt(id_vars=['E_lab', 'E*/MeV'], var_name='isotope', value_name='CS')
    dfm10 = dfm10_mb
    df10 = df10_mb
    
    mb = 1
    ub = 1e3
    nb = 1e6

    scale = mb
    dfm10['CS'] = dfm10_mb['CS']*scale
    sigr_dict[sigr_string] = dfm10   
    
    
colors = ['red', 'limegreen', 'blue', 'gold', 'black', 'turquoise', 'magenta', 'grey', 'olive', 'brown']
plot_from = 5
plot_to = 8
# labels = list(df10.columns[plot_from:plot_to])

fig, ax = plt.subplots()
new_labels = []
linestyles = ['-', ':', '-.', '--', (0, (2, 4))]

colors=colors[0:len(list(sigr_dict))]
for i in range(len(list(sigr_dict))):
    sigr = list(sigr_dict)[i]
    sns.lineplot(x='E_lab', y='CS', hue='isotope', data=sigr_dict[sigr][sigr_dict[sigr]['isotope'].isin(list(df10.columns[plot_from:plot_to]))], label='x',  ls=linestyles[i], legend=False)
    for label in labels:
        new_labels.append(label + ' - ' + sigr)

plt.semilogy()
plt.ylim(1e-3, 2e1)
plt.xlim(140, 230)
if scale == mb:
    plt.ylabel('$\sigma$ [mb]')
elif scale == ub:
    plt.ylabel('$\sigma$ [$\mu$b]')
elif scale == nb:
    plt.ylabel('$\sigma$ [nb]')
plt.title('Comparison of barrier fluctuation (sigr0) values')
plt.xlabel('$E_{proj}$ [MeV]')
h, l = ax.get_legend_handles_labels()
ax.legend(handles=h,labels=new_labels, loc='upper right')

plt.savefig(proj + '_' + trg + '_' + cn + '_' + ch + '_' + 'IFUS10' + '_barfac' + f'{bf:.2f}' + '_sigr-comparison_1.5-2.5-3.3-5.0.pdf')
plt.savefig(proj + '_' + trg + '_' + cn + '_' + ch + '_' + 'IFUS10' + '_barfac' + f'{bf:.2f}' + '_sigr-comparison_1.5-2.5-3.3-5.0.png')
```
