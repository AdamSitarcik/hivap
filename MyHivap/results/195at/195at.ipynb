## Plotting

```python
colors = ['red', 'limegreen', 'blue', 'gold', 'black', 'turquoise', 'magenta', 'grey', 'olive', 'brown']

plot_from_ship = 6
plot_from_ritu = 5

# frame = 'E*/MeV'
frame = 'E_lab'

fig, ax = plt.subplots(figsize=(12,8))
sns.lineplot(x=frame, y='CS', hue='isotope', data=dfm10_ship[dfm10_ship['isotope'].isin(list(df10_ship.columns[plot_from_ship:plot_from_ship+1]))], palette=[colors[0]])
sns.lineplot(x=frame, y='CS', hue='isotope', data=dfm0_ship[dfm0_ship['isotope'].isin(list(df0_ship.columns[plot_from_ship:plot_from_ship+1]))], palette=[colors[0]], ls=':', legend=False)

sns.lineplot(x=frame, y='CS', hue='isotope', data=dfm10_ritu[dfm10_ritu['isotope'].isin(list(df10_ritu.columns[plot_from_ritu:plot_from_ritu+1]))], palette=[colors[1]])
sns.lineplot(x=frame, y='CS', hue='isotope', data=dfm0_ritu[dfm0_ritu['isotope'].isin(list(df0_ritu.columns[plot_from_ritu:plot_from_ritu+1]))], palette=[colors[1]], ls=':', legend=False)

plt.semilogy()
plt.ylim(1e-4, 1e-1)
if frame == 'E_lab':
    plt.xlabel('$E_{proj}$ [MeV]')
    plt.xlim(220, 280)
elif frame == 'E*/MeV':
    plt.xlabel('$E*$ [MeV]')
    plt.xlim(30, 65)
plt.suptitle('CN: ' + cn + '*, ' + proj_ship + ' + ' + trg_ship + ' vs. ' + proj_ritu + ' + ' + trg_ritu)# + '; BF = ' + str(bf) + ' $\pm$ ' + str(bf_diff) + ', sigr0 = ' + str(sigr))#, va='top')
plt.title('BF = ' + str(bf) + ' $\pm$ ' + str(bf_diff) + ', sigr0 = ' + str(sigr) + ', V0 = 40 MeV', va='bottom', fontsize='small')
if scale == mb:
    plt.ylabel('$\sigma$ [mb]')
elif scale == ub:
    plt.ylabel('$\sigma$ [$\mu$b]')
elif scale == nb:
    plt.ylabel('$\sigma$ [nb]')
plt.legend(loc='upper right')

exp_ship = 40*1e-6*scale
exp_ritu = 300*1e-9*scale
        
if frame == 'E*/MeV':
    dfm_frame = 1
    E_exp_ship = 39
    E_exp_ritu = 49
    en_frame = 'Eexc'
elif frame == 'E_lab':
    dfm_frame = 0
    E_exp_ship = 230
    E_exp_ritu = 260
    en_frame = 'Elab'

(_, caps_ship, _) = plt.errorbar(E_exp_ship, exp_ship, yerr=exp_ship*0.25, marker='.', capsize=5, color=colors[0], ls='none', markersize=10)
(_, caps_ritu, _) = plt.errorbar(E_exp_ritu, exp_ritu, yerr=0, marker='.', capsize=5, color=colors[1], ls='none', markersize=10)

alpha_value = 0.1

for i in range(plot_from_ship, plot_from_ship+1):
    plt.fill_between(df10_ship.iloc[:, dfm_frame],  df10_ship.iloc[:, i], bf_low_10_ship.iloc[:, i], color=colors[0], alpha = alpha_value)
    plt.fill_between(df10_ship.iloc[:, dfm_frame],  df10_ship.iloc[:, i], bf_high_10_ship.iloc[:, i], color=colors[0], alpha = alpha_value)
    # plt.fill_between(df0_ship.iloc[:, 0],  df0_ship.iloc[:, i], bf_low_0_ship.iloc[:, i], color=colors[i-plot_from], alpha = alpha_value)
    # plt.fill_between(df0_ship.iloc[:, 0],  df0_ship.iloc[:, i], bf_high_0_ship.iloc[:, i], color=colors[i-plot_from], alpha = alpha_value)
    
for i in range(plot_from_ritu, plot_from_ritu+1):
    plt.fill_between(df10_ritu.iloc[:, dfm_frame],  df10_ritu.iloc[:, i], bf_low_10_ritu.iloc[:, i], color=colors[1], alpha = alpha_value)
    plt.fill_between(df10_ritu.iloc[:, dfm_frame],  df10_ritu.iloc[:, i], bf_high_10_ritu.iloc[:, i], color=colors[1], alpha = alpha_value)
    
for cap in caps_ship:
    cap.set_markeredgewidth(1)
for cap in caps_ritu:
    cap.set_markeredgewidth(1)
    
plt.rcParams.update({'font.size': 15})

plt.savefig('195At_SHIP_vs_RITU' + '_barfac' + str(bf) + '_sigr' + str(sigr) + '_' + en_frame + '.png')
plt.savefig('195At_SHIP_vs_RITU' + '_barfac' + str(bf) + '_sigr' + str(sigr) + '_' + en_frame + '.pdf')
```
