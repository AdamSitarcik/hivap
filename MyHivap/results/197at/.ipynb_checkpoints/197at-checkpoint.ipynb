```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
proj = '56Fe'
trg = '141Pr'
cn = '197At'
bf = 0.66
ch = 'xn'

file_name10 = proj + '_' + trg + '_' + cn + '_' + ch + '_' + 'IFUS10' + '_' + 'barfac' + str(bf) + '.dat'
file_name0 = proj + '_' + trg + '_' + cn + '_' + ch + '_' + 'IFUS0' + '_' + 'barfac' + str(bf) + '.dat'
print(file_name10, file_name0)
```

```python
df10 = pd.read_csv('data/' + file_name10, sep='\t')
df0 = pd.read_csv('data/' + file_name0, sep='\t')
```

```python
df10.head()
```

```python
df0.head()
```

```python
dfm10_mb = df10.melt(id_vars=['E_lab', 'E*/MeV'], var_name='isotope', value_name='CS')
dfm0_mb = df0.melt(id_vars=['E_lab', 'E*/MeV'], var_name='isotope', value_name='CS')

#scale the data to either nb, ub or whatever
dfm10 = dfm10_mb
dfm0 = dfm0_mb

dfm10['CS'] = dfm10_mb['CS']*1000000
dfm0['CS'] = dfm0_mb['CS']*1E6
dfm10_mb.head()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.lineplot(x='E*/MeV', y='CS', hue='isotope', data=dfm10[dfm10['isotope'].isin(list(df10.columns[3:-2]))])
plt.semilogy()
plt.ylim(1e-0, 1e3)
# plt.xlim(right=50)
plt.xlim(20, 80)
plt.title(proj + ' + ' + trg + ' -> ' + cn + ', BF = ' + str(bf))
plt.ylabel('$\sigma$ [nb]')
plt.xlabel('$E^*$ [MeV]')
plt.legend(loc='upper right')

sns.lineplot(x='E*/MeV', y='CS', hue='isotope', data=dfm0, linestyle='--', legend=None)

plt.savefig(proj + '_' + trg + '_' + cn + '_' + ch + '_' + '_' + 'barfac' + str(bf) + '.pdf')
```
