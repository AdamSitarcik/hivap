```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a_cn = np.arange(195, 207, 1)
bf = [0.6, np.nan, 0.66, 0.69, np.nan, np.nan, 0.65, 0.66, 0.68, np.nan, 0.81,  0.82]
```

```python
fig, ax = plt.subplots(figsize=(12,8))

plt.errorbar(a_cn, bf, yerr=0.0, marker='.', ls='none', markersize=20, lw=3)
plt.xticks(np.arange(min(a_cn), max(a_cn)+1, 1))
plt.yticks(np.arange(min(bf), max(bf)+0.01, 0.02))
plt.title('BF values for At* CN')
plt.ylabel('BF', fontsize=25)
plt.xlabel('$A_{CN}$', fontsize=25)
plt.grid()

plt.rcParams.update({'font.size': 15})
```
